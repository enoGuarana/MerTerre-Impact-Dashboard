```python
# 1. INSTALA√á√ÉO E IMPORTA√á√ÉO DE BIBLIOTECAS

import pandas as pd
import folium # Para o Mapa
import plotly.express as px # Para os Gr√°ficoS
from IPython.display import display, HTML

# T√≠tulo da Apresenta√ß√£o
display(HTML("<h1>üåä MerTerre: Dashboard de Impacto (Prot√≥tipo)</h1>"))

# ---------------------------------------------------------
# 2. CRIANDO DADOS FICT√çCIOS (SIMULA√á√ÉO)
# Focados em Marselha e nas Calanques
# ---------------------------------------------------------

dados = {
    'Local': ['Calanque de Sormiou', 'Plage du Prado', 'Vieux Port', 'Calanque de Morgiou', 'L\'Estaque'],
    'Latitude': [43.2104, 43.2556, 43.2951, 43.2125, 43.3630],
    'Longitude': [5.4194, 5.3730, 5.3744, 5.4428, 5.3190],
    'Kg de Lixo': [120, 350, 410, 95, 210],
    'Status': ['Adotado ‚úÖ', 'Precisa de A√ß√£o ‚ö†Ô∏è', 'A√ß√£o Agendada üìÖ', 'Adotado ‚úÖ', 'Precisa de A√ß√£o ‚ö†Ô∏è'],
    'Volunt√°rios': [15, 40, 55, 12, 25],
    'Tipo Principal': ['Pl√°stico', 'Bitucas', 'Pl√°stico', 'Vidro', 'Metal']
}

df = pd.DataFrame(dados)

# ---------------------------------------------------------
# 3. VISUALIZA√á√ÉO 1: MAPA INTERATIVO (ADOPT'1 SPOT)
# ---------------------------------------------------------
display(HTML("<h3>üìç Mapeamento: Projeto Adopt'1 Spot</h3>"))

# Cria o mapa centrado em Marselha
mapa = folium.Map(location=[43.2965, 5.3698], zoom_start=11)

# Adiciona os pontos no mapa
for i, row in df.iterrows():
    # Define cor baseada no status
    cor = 'green' if 'Adotado' in row['Status'] else 'red'
    if 'Agendada' in row['Status']: cor = 'orange'

    # Cria o popup com informa√ß√µes
    info = f"<b>{row['Local']}</b><br>Status: {row['Status']}<br>Lixo: {row['Kg de Lixo']}kg"

    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=info,
        icon=folium.Icon(color=cor, icon='info-sign')
    ).add_to(mapa)

# Mostra o mapa na tela
display(mapa)

# ---------------------------------------------------------
# 4. VISUALIZA√á√ÉO 2: GR√ÅFICOS (Z√âRO D√âCHET)
# ---------------------------------------------------------
display(HTML("<h3>üìä An√°lise de Res√≠duos: Z√©ro D√©chet Sauvage</h3>"))

# Gr√°fico de Barras: Kg por Local
fig1 = px.bar(df, x='Local', y='Kg de Lixo',
             color='Tipo Principal',
             title='Total de Lixo Coletado por Local e Tipo',
             text='Kg de Lixo')
fig1.update_layout(height=400)
fig1.show()

# Gr√°fico de Pizza: Distribui√ß√£o de Status
fig2 = px.pie(df, names='Status', title='Status das Zonas de Ado√ß√£o', hole=0.4)
fig2.update_layout(height=400)
fig2.show()
```
